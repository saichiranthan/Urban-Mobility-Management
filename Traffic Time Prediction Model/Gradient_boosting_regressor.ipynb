```python
# Initialize the models
gbr_normal = GradientBoostingRegressor(random_state=42)
gbr_peak = GradientBoostingRegressor(random_state=42)
```

```python
# Train the models
y_train = df.loc[X_train.index, 'duration']
y_test = df.loc[X_test.index, 'duration']
gbr_normal.fit(X_train, y_train)
y_train = df.loc[X_train.index, 'duration_in_traffic']
y_test = df.loc[X_test.index, 'duration_in_traffic']
gbr_peak.fit(X_train, y_train)
```

```python
# Make predictions
y_pred_normal = gbr_normal.predict(X_test)
y_pred_peak = gbr_peak.predict(X_test)
```

```python
# Evaluate the models
mse_normal = mean_squared_error(y_test, y_pred_normal)
r2_normal = r2_score(y_test, y_pred_normal)
mse_peak = mean_squared_error(y_test, y_pred_peak)
r2_peak = r2_score(y_test, y_pred_peak)
```
